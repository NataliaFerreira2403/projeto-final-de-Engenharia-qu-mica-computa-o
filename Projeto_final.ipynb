```python
#Listas
#contém a operação utilizada na planta.
#seu conteúdo: nome,tipo,temperatura,pressão,vazão,pH.
operações_unitárias=[
  ["Reator 1","reator", 70, 3.5, 20 , 11],
  ["Tanque A","Tanque", 85, 2, 20, 8],
  ["trocador H1","Trocador",120,2.3,5,"-"],
  ["coluna 2B","Coluna", 60,0.8,2,5]]

Faixas_de_segurança = {
    "reator": {
        "temp_min": 55,
        "temp_max": 90,
        "psi_min" : 2.8,
        "psi_max" : 15,
        "vazão_min": 18,
        "vazão_max": 35,
        "pH_min": 6.5,
        "pH_max": 14
    },
    "Tanque": {
        "temp_min": 60,
        "temp_max": 100,
        "psi_min": 1.5,
        "psi_max": 4,
        "vazão_min": 10,
        "vazão_max": 30,
        "pH_min": 6.5,
        "pH_max": 11
    },
    "Trocador": {
        "temp_min": 100,
        "temp_max": 300,
        "psi_min": 1.8,
        "psi_max": 4.5,
        "vazão_min": 3,
        "vazão_max": 10
    },
    "Coluna": {
        "temp_min": 40,
        "temp_max": 90,
        "psi_min": 0.7,
        "psi_max": 1,
        "vazão_min": 1,
        "vazão_max": 2.5,
        "pH_min": 4.5,
        "pH_max": 7
    }
}

def dentro_faixa(operação_unitária):
  nome, tipo, temperatura, pressão, vazão, pH = operação_unitária
  regras = Faixas_de_segurança.get(tipo, None)
  if regras is None:
    return True #se não tiver regra, consideramos ok
  if not (regras["temp_min"] <= temperatura <= regras["temp_max"]):
    return False
  if not (regras["psi_min"] <= pressão <= regras["psi_max"]):
    return False
  if not (regras["vazão_min"] <= vazão <= regras["vazão_max"]):
    return False
  if pH != "-" and "pH_min"in regras:
      if not (regras["pH_min"] <= float(pH) <= regras["pH_max"]):
          return False
  return True
for u in operações_unitárias:
  print(u[0], "dentro da faixa?", dentro_faixa(u))

def registrar_medicao(entrada):
  partes = entrada.split(";")
  nome = partes[0].strip()

  medidas = {}
  for pedaco in partes[1:]:
      pedaco = pedaco.strip()
      if "=" in pedaco:
          k, v = pedaco.split("=")
          medidas[k.strip().lower()] = v.strip()
  for u in operações_unitárias:
      if u [0] == nome:
        if "T" in medidas:
          u[2] = float(medidas["T"])
        if "P" in medidas:
          u[3] = float(medidas["P"])
        if "V" in medidas:
          u[4] = float(medidas["V"])
        if "pH" in medidas:
          u[5] = float(medidas["pH"])
        break
```
